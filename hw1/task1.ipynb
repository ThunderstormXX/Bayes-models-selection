```python
import random
import numpy as np
from matplotlib import pyplot as plt
```

```python
p1 = 0.5
p2 = 0.9 #oracle
N_members = 100
def get_members(P_oracle) :
    return np.random.binomial(n=1, p=P_oracle, size=N_members)
def P(member):
    if member == 1 :
        return p2
    else :
        return p1
def prior_oracle_by_competition(N_sampling ,k , P_oracle) :
    counter = 0
    for _ in range(N_sampling) :
        members = get_members(P_oracle)
        # Competition
        top_score = 0
        top_mem = None
        for mem in members :
            score = np.sum(np.random.binomial(n=1, p=P(mem), size=k))
            if score > top_score :
                top_score = score 
                top_mem = mem
        counter = counter + top_mem
    return counter/N_sampling
```

#### Диапазон k $\in$ [1,100]

```python
P_oracle =0.0001
N_sampling = 10000
k_array = np.arange(1,100,10)

priors = []
for k in k_array :
    priors.append(prior_oracle_by_competition(N_sampling,k ,P_oracle))

plt.plot(k_array ,priors)
plt.show()
```

#### Диапазон k $\in$ [100,1000]

```python
P_oracle =0.0001
N_sampling = 10000
k_array = np.arange(99,1000,50) 

print(k_array)

priors = []
for k in k_array :
    priors.append(prior_oracle_by_competition(N_sampling,k ,P_oracle))

plt.plot(k_array ,priors)
plt.show()
```
